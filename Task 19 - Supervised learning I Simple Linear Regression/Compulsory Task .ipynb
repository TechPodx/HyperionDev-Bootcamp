```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
insu_data = pd.read_csv("insurance.csv" , usecols = ["age" , "charges"])
insu_data.head()
```

```python
# Selecting rows to create scatter plot
x = insu_data.iloc[ : , 0].values.reshape(-1 , 1 )
y = insu_data.iloc[ : , -1].values.reshape(-1 , 1)

plt.scatter(x , y , color = "g")
plt.xlabel("Age")
plt.ylabel("Insurance Charges")
plt.title("Age and Insurance Charges Comparison")
plt.show()
```

```python
# Fit the linear regression model to the data
fit_regression = LinearRegression()
fit_regression.fit(x.reshape(-1 , 1) , y)

# Make predictions
x_uknw = np.array([[70]])
y_predict = fit_regression.predict(x_uknw.reshape(-1 , 1))
print(f"Insurance charges for a person age of {x_uknw} is {y_predict}")
```

```python
# With best fit line
plt.scatter(x , y , color = "g")
plt.plot(x , fit_regression.predict(x) , color = "b")
plt.xlabel("Age")
plt.ylabel("Insurance Charges")
plt.title("Age and Insurance Charges Comparison")
plt.show()
```
